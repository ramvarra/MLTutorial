```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (12,8)
%matplotlib inline
```

### Normal Distribution

```python

σ =  2
μ = 6
N = 1000

v = np.random.normal(σ, μ, size=N)
fig, axs= plt.subplots(nrows=1, ncols=2, figsize=(12,5))
axs[0].plot(v, '.')
sns.distplot(v, ax=axs[1]);
```

```python
v.mean(), v.std()
```

### Poisson Distribution
math:: $f(k; \lambda)=\frac{\lambda^k e^{-\lambda}}{k!}$

roots: $\frac{-b}{2a}$


```python
λ = 1
N = 100
v = np.random.poisson(λ, size=N)
fig, axs= plt.subplots(nrows=1, ncols=2, figsize=(12,5))
axs[0].plot(v, '.')
sns.distplot(v, ax=axs[1]);
```

### Student T Distribution

```python
df = 5
N = 1000
v = np.random.standard_t(df, size=N)
fig, axs= plt.subplots(nrows=1, ncols=2, figsize=(12,5))
axs[0].plot(v, '.')
sns.distplot(v, ax=axs[1]);
```
