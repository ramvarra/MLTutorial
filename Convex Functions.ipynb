```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Convex Funtion $2x^2+3y^2-5$

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
xx, yy = np.meshgrid(x, y)
z = 2*np.square(xx) + 3*np.square(yy) - 5
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'$Convex \  Function \  2x^2+3y^2-5$', fontsize=20);
```

### Non Convex function $log\left(\frac{x}{y}\right)$

```python
x = np.linspace(1, 100, 100)
y = np.linspace(1, 100, 100)
xx, yy = np.meshgrid(x, y)
z = np.log(xx/yy)
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'Non Convex Function: $log\left(\frac{x}{y}\right)$', fontsize=20)
```

###  Minima and Maxima $xe^{(-x^2-y^2)}$

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xx, yy = np.meshgrid(x, y)
z = xx * np.exp(-1*np.square(xx) - np.square(yy))
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'Minima and Maxima Function: $xe^{(-x^2-y^2)}$', fontsize=20)
```
